# SRISHTI PANDEY - COVID 19 DATA ANALYSIS PROJECT

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
corona_dataset_csv = pd.read_csv("covid19_Confirmed_dataset.csv")
corona_dataset_csv.head()
```

```python
corona_dataset_csv.shape
```

```python
corona_dataset_csv.drop(["Lat" ,"Long"], axis = 1,inplace = True )
```

```python
corona_dataset_csv.head()
```

#Aggregating State Col with Country Col. 

```python
corona_dataset_aggregated = corona_dataset_csv.groupby("Country/Region").sum()
```

```python
corona_dataset_aggregated.head(10)
```

```python
corona_dataset_aggregated.shape    
```

#Visualizing data with country name 

```python
corona_dataset_aggregated.loc["China"].plot()
corona_dataset_aggregated.loc["India"].plot()
corona_dataset_aggregated.loc["Italy"].plot()
plt.legend()
```

```python
corona_dataset_aggregated.loc["China"][:3].plot()
corona_dataset_aggregated.loc["India"]["4/21/20":"4/25/20"].plot()
plt.legend()
```

#plotting maximum hike in cases

```python
corona_dataset_aggregated.loc["China"].diff().plot()
```

#Calculating maximum cases discovered in single day

```python
corona_dataset_aggregated.loc["China"].diff().max()
```

```python
corona_dataset_aggregated.loc["Italy"].diff().max()
```

#Calculating maximum Covid cases for all countries

```python
countries = list(corona_dataset_aggregated.index)
max_infection_rates = []
for c in countries:
    max_infection_rates.append(corona_dataset_aggregated.loc[c].diff().max())
corona_dataset_aggregated["Total COVID 19 Cases"] = max_infection_rates
corona_dataset_aggregated.head()
```

#Copying final data to new dataframe

```python
corona_data = pd.DataFrame(corona_dataset_aggregated["Total COVID 19 Cases"])
corona_data.head(15)
```

#Analysing data after COVID 19

```python
happiness_report_csv = pd.read_csv("worldwide_happiness_report.csv")
happiness_report_csv.head()
```

```python
Useless_cols = ["Social support", "Generosity"]
happiness_report_csv.drop(Useless_cols,axis = 1, inplace = True)
```

```python
happiness_report_csv.head()
```

```python
happiness_report_csv.drop("Overall rank", axis = 1, inplace = True)
```

```python
happiness_report_csv.head()
```

```python
happiness_report_csv.set_index("Country or region", inplace = True)
happiness_report_csv.head()
```
